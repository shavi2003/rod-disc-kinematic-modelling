## 1. Assumptions per run

- The wall angle \(\theta\) is **fixed** during each experiment run. You repeat for different \(\theta\).
- The disc of radius \(R\) **rolls without slipping** on the ground for the baseline model.
- The rod \(AB\) is **rigid**. Point \(B\) is pinned to the disc; point \(A\) **slides** along the guide line at angle \(\theta\).
- We use **planar rigid-body kinematics** only (positions, velocities, accelerations). No forces/torques.


\
## 2. What we extract from video

Track the coloured markers to obtain positions as functions of time (after converting pixels→cm and frames→s):
- \( \mathbf{r}_O(t) = (x_O(t),\,0) \): disc centre.
- \( \mathbf{r}_B(t) \): pin on disc.
- \( \mathbf{r}_A(t) \): slider on the wall.

From these we compute:
- Disc angle: \( \gamma(t) = \operatorname{atan2}\!\big(y_B-y_O,\; x_B-x_O\big) \).
- Rod angle: \( \alpha(t) = \operatorname{atan2}\!\big(y_B-y_A,\; x_B-x_A\big) \).
- Slider coordinate along the wall: \( s(t) = \mathbf{r}_A \cdot (\cos\theta,\,\sin\theta) = x_A\cos\theta + y_A\sin\theta \).

Apply smoothing and numerical differentiation to obtain \( \dot{\gamma}, \ddot{\gamma}, \dot{\alpha}, \ddot{\alpha}, \dot{s}, \ddot{s} \).


\
## 3. Geometry (positions)

Let \( \mathbf{i} \) be horizontal and \( \mathbf{j} \) vertical. With disc radius \(R\) and rod length \(L\):
\[
\begin{aligned}
\mathbf{r}_O &= (x_O,\,0), \\[4pt]
\mathbf{r}_B &= \mathbf{r}_O + R\,[\cos\gamma,\,\sin\gamma], \\[4pt]
\mathbf{r}_A &= s\,[\cos\theta,\,\sin\theta], \\[4pt]
\mathbf{r}_{B/A} &= \mathbf{r}_B - \mathbf{r}_A = L\,[\cos\alpha,\,\sin\alpha].
\end{aligned}
\]


\
## 4. Rolling (no-slip) constraint

Pure rolling imposes zero velocity at the contact with the ground, giving
\[
\boxed{ \ \dot{x}_O \;=\; R\,\dot{\gamma} \ }.
\]
Equivalently \( \mathbf{v}_O = (\dot{x}_O,\,0) = (R\dot{\gamma},\,0) \) and \( \mathbf{a}_O = (R\ddot{\gamma},\,0) \).
This is the **baseline** constraint to eliminate one unknown and to validate with data.


\
## 5. Velocity relations (rigid-body kinematics)

For any rigid body in planar motion with angular velocity \( \boldsymbol{\omega} = \omega\,\mathbf{k} \),
\[
\boxed{ \ \mathbf{v}_B \;=\; \mathbf{v}_A \;+\; \boldsymbol{\omega}\times \mathbf{r}_{B/A} \ }.
\]

Apply to **disc** (points \(O\to B\)):
\[
\boldsymbol{\omega}_2 = \dot{\gamma}\,\mathbf{k},\qquad \mathbf{r}_{B/O} = R[\cos\gamma,\,\sin\gamma].
\]
Hence
\[
\boxed{\ \mathbf{v}_B^{(\mathrm{disc})} = \mathbf{v}_O + \boldsymbol{\omega}_2\times \mathbf{r}_{B/O}
= \big(\dot{x}_O - R\dot{\gamma}\sin\gamma,\;\; R\dot{\gamma}\cos\gamma\big). \ }
\]

Apply to **rod** (points \(A\to B\)):
\[
\boldsymbol{\omega}_1 = \dot{\alpha}\,\mathbf{k},\qquad \mathbf{r}_{B/A} = L[\cos\alpha,\,\sin\alpha].
\]
Hence
\[
\boxed{\ \mathbf{v}_B^{(\mathrm{rod})} = \mathbf{v}_A + \boldsymbol{\omega}_1\times \mathbf{r}_{B/A}
= \big(\dot{s}\cos\theta - L\dot{\alpha}\sin\alpha,\;\; \dot{s}\sin\theta + L\dot{\alpha}\cos\alpha\big). \ }
\]

Equate components of \( \mathbf{v}_B^{(\mathrm{disc})} \) and \( \mathbf{v}_B^{(\mathrm{rod})} \) to obtain a **2×2 system** for \( \dot{s} \) and \( \dot{\alpha} \):
\[
\begin{bmatrix}
\cos\theta & -L\sin\alpha \\
\sin\theta & \phantom{-}L\cos\alpha
\end{bmatrix}
\begin{bmatrix}
\dot{s}\\[2pt]\dot{\alpha}
\end{bmatrix}
=
\begin{bmatrix}
\dot{x}_O - R\dot{\gamma}\sin\gamma \\[2pt]
R\dot{\gamma}\cos\gamma
\end{bmatrix}.
\]

With **no slip** \( (\dot{x}_O=R\dot{\gamma}) \) this becomes an explicit prediction for \( \dot{s} \) and \( \dot{\alpha} \) in terms of \(\gamma,\dot{\gamma},\alpha,\theta\).


\
## 6. Acceleration relations (rigid-body kinematics)

For planar motion with angular acceleration \( \boldsymbol{\varepsilon} = \dot{\omega}\,\mathbf{k} \),
\[
\boxed{\ \mathbf{a}_B \;=\; \mathbf{a}_A \;+\; \boldsymbol{\varepsilon}\times \mathbf{r}_{B/A} \;-\; \omega^2\,\mathbf{r}_{B/A} \ }.
\]

**Disc**:
\[
\boldsymbol{\varepsilon}_2 = \ddot{\gamma}\,\mathbf{k},\qquad \mathbf{a}_O = (R\ddot{\gamma},\,0),\qquad \mathbf{r}_{B/O}=R[\cos\gamma,\,\sin\gamma].
\]
Therefore
\[
\boxed{
\begin{aligned}
a_{Bx}^{(\mathrm{disc})} &= R\!\left[\ddot{\gamma}(1-\sin\gamma) - \dot{\gamma}^{\,2}\cos\gamma\right], \\[4pt]
a_{By}^{(\mathrm{disc})} &= R\!\left[\ddot{\gamma}\cos\gamma - \dot{\gamma}^{\,2}\sin\gamma\right].
\end{aligned}}
\]

**Rod**:
\[
\boldsymbol{\varepsilon}_1 = \ddot{\alpha}\,\mathbf{k},\qquad \mathbf{a}_A = \ddot{s}\,[\cos\theta,\,\sin\theta],\qquad \mathbf{r}_{B/A}=L[\cos\alpha,\,\sin\alpha].
\]
Hence
\[
\boxed{
\mathbf{a}_B^{(\mathrm{rod})} =
\big(\ddot{s}\cos\theta - L\ddot{\alpha}\sin\alpha - L\dot{\alpha}^{\,2}\cos\alpha,\;\;
    \ddot{s}\sin\theta + L\ddot{\alpha}\cos\alpha - L\dot{\alpha}^{\,2}\sin\alpha\big).}
\]

Equate components to solve the **2×2 system** for \( \ddot{s} \) and \( \ddot{\alpha} \):
\[
\begin{bmatrix}
\cos\theta & -L\sin\alpha \\
\sin\theta & \phantom{-}L\cos\alpha
\end{bmatrix}
\begin{bmatrix}
\ddot{s}\\[2pt]\ddot{\alpha}
\end{bmatrix}
=
\begin{bmatrix}
a_{Bx}^{(\mathrm{disc})}\\[2pt]
a_{By}^{(\mathrm{disc})}
\end{bmatrix}
+
L\dot{\alpha}^{\,2}
\begin{bmatrix}
\cos\alpha\\[2pt]\sin\alpha
\end{bmatrix}.
\]

This yields acceleration-level predictions \( (\ddot{s},\,\ddot{\alpha}) \) in terms of measured \( \gamma,\dot{\gamma},\ddot{\gamma} \) and geometry \( \alpha,\theta \).


\
## 7. Instantaneous Centre of Zero Velocity (ICZV) of the rod

For planar motion, the point \(P\) on the rod with zero velocity at that instant satisfies
\[
\boxed{ \ \mathbf{r}_{AP} \;=\; \frac{\boldsymbol{\omega}_1 \times \mathbf{v}_A}{\lVert \boldsymbol{\omega}_1 \rVert^{2}} \ }.
\]
Geometrically, the ICZV lies along the line through \(A\) perpendicular to \( \mathbf{v}_A \). Plotting \(P\) over time provides a visual validation of the kinematics.


\
## 8. Practical sequence (what you do and why)

1. **Capture & track** \(O,A,B\) to obtain \( \mathbf{r}_O(t), \mathbf{r}_A(t), \mathbf{r}_B(t) \).
2. **Compute angles & slider**: \( \gamma(t), \alpha(t), s(t) \). Smooth and differentiate to get rates.
3. **Check no-slip**: compare \( \dot{x}_O \) from data with \( R\dot{\gamma} \). Plot the residual; small error supports the baseline assumption.
4. **Velocity-level test**: use the 2×2 linear system to predict \( \dot{s}, \dot{\alpha} \) and compare with differentiated data.
5. **Acceleration-level test**: build \( \mathbf{a}_B^{(\mathrm{disc})} \) from \( \gamma,\dot{\gamma},\ddot{\gamma} \), solve for \( \ddot{s}, \ddot{\alpha} \) and compare with data.
6. **ICZV**: compute and overlay the ICZV on frames for geometric intuition.
7. **Repeat for other \( \theta \)**: report how metrics (e.g., peak \( |\dot{\alpha}| \), RMS residuals) vary with \( \theta \).


\
## 9. What is theoretical vs experimental (and how we compare)

- **Theoretical (model)**: the rigid-body relations above (velocity and acceleration equalities at \(B\)), the rolling constraint, and the ICZV construction.
- **Experimental (data)**: tracked positions \( \mathbf{r}_O,\mathbf{r}_A,\mathbf{r}_B \) and their derived \( \gamma,\alpha,s \) with time-derivatives.

**Comparisons to show:**
- Component-wise differences \( \mathbf{v}_B^{(\mathrm{disc})} - \mathbf{v}_B^{(\mathrm{rod})} \) and \( \mathbf{a}_B^{(\mathrm{disc})} - \mathbf{a}_B^{(\mathrm{rod})} \) vs time.
- \( \dot{x}_O \) vs \( R\dot{\gamma} \) (no-slip check) and RMS error.
- Predicted vs measured \( (\dot{s},\dot{\alpha}) \) and \( (\ddot{s},\ddot{\alpha}) \) with residual plots.
